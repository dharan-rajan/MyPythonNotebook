```python
class StudentMarkSheet:
    subjects = ('Language', 'English', 'Maths', 'Science', 'Social')
    def __init__(self, marks):
        self.marks = marks

    def __str__(self):
        markStr = "Marks:"
        for subject, mark in zip(self.subjects, self.marks):
            markStr += f"\n{subject:<10} = {mark}"
        return markStr

    def __repr__(self):
        """Unambigous - just for debugging or eval"""
        return StudentMarkSheet.__name__+ f"({self.marks})"

    @property
    def total(self):
        return sum(self.marks)

    def __abs__(self):
        return self.total

    def __bool__(self):
        return bool(self.marks) and len(self.marks) != 0

    def __eq__(self, other):
        return self.marks == other.marks

    def __getitem__(self, index): 
        # Becoming virtual subclass of Iterable
        return self.marks[index]

    def __sub__(self, other):
        return self.total - other.total

    def __iter__(self):
        # Alernate idea- return generator
        self.index=-1
        return self

    def __next__(self):
        if self.index < len(self.marks)-1:
            self.index += 1
            return self.marks[self.index]
        raise StopIteration

         
marks = StudentMarkSheet([78, 84, 85, 80, 90])

# Representation
# print(marks)
# print(repr(marks))

# Dictionaries - Class and Object
# print(StudentMarkSheet.__dict__)
# print(marks.__dict__)

# Changing Dictionaries
# marks.__dict__['minMark'] = 50
# print(marks.__dict__)
# StudentMarkSheet.__dict__['passMark'] = 50 - this is not possible readOnly

# Property and bool
# marks.total
# print(abs(marks))
# bool(marks)

# Testing eq method with different marks
# marks1 = StudentMarkSheet([78, 84, 85, 80, 91])
# marks == marks1 

# Iterable
# for mark in marks:
#     print(mark)
# from collections.abc import Iterable
# print(isinstance(marks, Iterable)) # False
# for i in marks:
#     print(i)

# Difference of marks
# marks2 = StudentMarkSheet([80, 84, 95, 89, 80])
# print(marks - marks2)
```

```python
class Vehicle(ABC):
    @abc.abstractmethod
    def start(self):
        print("Vehicle is starting")

    @abc.abstractmethod
    def stop(self):
        print("Vehicle is stopped")

class Car(Vehicle):
    wheels = 4
    def __init__(self):
        pass

    def start(self):
        print("Car is starting")

    def stop(self):
        print("Car is starting")
        
    def description(cls):
        print(f"A vehicle with {cls.wheels} wheels")

    def __infoManual(self):
        print("Information manual")

car = Car()
car.description()
car.start()

# Private methods are hidden like this 
# and not pythonic way to call this way
car._Car__infoManual() 
```
